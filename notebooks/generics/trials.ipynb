```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data
np.random.seed(0)
data = np.random.rand(100, 2)

# Convert to polar coordinates
r = np.sqrt(data[:, 0]**2 + data[:, 1]**2)
theta = np.arctan2(data[:, 1], data[:, 0])


# Box-Muller transform
u1 = np.random.rand(len(theta))
u2 = np.random.rand(len(theta))

# Generate normally distributed values
normal_theta = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)

# Plot the new theta values
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
c = ax.scatter(normal_theta, r, c=normal_theta, s=20, cmap='hsv', alpha=0.75)

# Set the radius limit to 1 (unit circle)
ax.set_rlim(0, 1)

plt.show()
```

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
np.random.seed(0)
r = np.random.rand(100)
theta = 2 * np.pi * np.random.rand(100)

# Convert to Cartesian coordinates for GMM
x = r * np.cos(theta)
y = r * np.sin(theta)
data = np.column_stack([x, y])

# Fit a Gaussian mixture model
gmm = GaussianMixture(n_components=2)
gmm.fit(data)

# Predict the cluster for each data point
labels = gmm.predict(data)

# Plot the data and the clusters
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
scatter = ax.scatter(theta, r, c=labels, s=20, cmap='viridis', alpha=0.75)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate data for the first cluster
r1 = np.random.normal(loc=0.3, scale=0.1, size=50)  # radius
theta1 = np.random.uniform(low=0, high=2*np.pi, size=50)  # angle
label1 = np.zeros(50)  # label

# Generate data for the second cluster
r2 = np.random.normal(loc=0.7, scale=0.1, size=50)  # radius
theta2 = np.random.uniform(low=0, high=2*np.pi, size=50)  # angle
label2 = np.ones(50)  # label

# Combine the data
r = np.concatenate((r1, r2))
theta = np.concatenate((theta1, theta2))
labels = np.concatenate((label1, label2))

# Plot the data
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.scatter(theta, r, c=labels, alpha=0.75)

plt.show()
```
